```python
def fetch_video_times(value):
    return daystrending[value]
```

```python
data2maxviews["times_in_trending"] = data2maxviews.video_id.apply(fetch_video_times)
data2maxviews.sample(5)
```

```python
data2_trend_video = data2maxviews.sort_values('times_in_trending', ascending=False).iloc[:25,:]
data2_trend_video
```

```python
#Top 20 videos with most no.of times in trending(Channel wise)

# Set the width and height of the figure
plt.figure(figsize=(20,10))

# Add title
plt.title("Channels vs no.of times in Trending")
plt.xlabel
plt.xticks(rotation=90)

sns.barplot(x=data2_trend_video.channelTitle, y=data2_trend_video.times_in_trending)
#sns.violinplot(x = df_trend_video_tt.channelTitle, y = df_minViews.DaysTakenToBeOnTrending )

# Add label for vertical axis
plt.ylabel("no.of trending")
```

```python
data2_Channel_Trending = data2maxviews.groupby("channelTitle")["times_in_trending"].sum().reset_index()
data2_Channel_Trending["Majority_Videos_Category_Type"] = data2maxviews.groupby('channelTitle')['Video_Category'].agg(pd.Series.mode).reset_index()["Video_Category"]
data2_Channel_Trending = data2_Channel_Trending.sort_values('times_in_trending', ascending=False).iloc[:30,:]
data2_Channel_Trending 
```

```python
# Set the width and height of the figure
plt.figure(figsize=(20,10))

# Add title
plt.title("Channel vs total no.of times in trending")
plt.xlabel
plt.xticks(rotation=90)

sns.barplot(x=data2_Channel_Trending.channelTitle, y=data2_Channel_Trending.times_in_trending)

# Add label for vertical axis
plt.ylabel("no.of times in trending")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(20,10))

# Add title
plt.title("Category vs total no.of times in trending")
plt.xlabel
plt.xticks(rotation=90)
Category_Trending = data2_Channel_Trending.groupby("Majority_Videos_Category_Type")["times_in_trending"].sum().reset_index()
sns.barplot(x=Category_Trending.Majority_Videos_Category_Type, y=Category_Trending.times_in_trending)

# Add label for vertical axis
plt.ylabel("no.of times in trending")
```

```python
data2minviews["JoinedTrending"] = pd.to_datetime(data2minviews.trending_date.str.split(',\s*').str[0]) #\s means matches "any whitespace character"
data2minviews["UploadedTime"] = pd.to_datetime(data2minviews.publishedAt.str.split(',\s*').str[0])
data2minviews["DaysTakenToBeOnTrending"] = (data2minviews["JoinedTrending"] - data2minviews["UploadedTime"]).dt.days
data2minviews['DaysTakenToBeOnTrending'] = data2minviews['DaysTakenToBeOnTrending'].apply(lambda x: 0 if x == -1 else x)
data2minviews.sample(10)
```

```python
#Top 20 videos with most no.of times in trending(Channel wise)

# Set the width and height of the figure
plt.figure(figsize=(20,10))

# Add title
plt.title("Average no.of days took to be in trending")
plt.xlabel
plt.xticks(rotation=90)

sns.barplot(x=data2minviews.Video_Category, y=data2minviews['DaysTakenToBeOnTrending'])
#sns.violinplot(x = df_minViews.Video_Category, y = df_minViews.DaysTakenToBeOnTrending )

# Add label for vertical axis
plt.ylabel("Avg no.of days")
```
